###1. dataset load, 전체적인 개형 살펴보기 

```python
df = pd.read_csv("dataset.csv")
df = pd.DataFrame(df)
```

```python
df
```

```python
#결측치 등 확인
df.info()

#연락처(전화번호) 외의 컬럼에는 결측치가 없는 것을 확인
```

전국구 데이터를 입력하기 위해 '연번'을 삭제하고 '소재지'를 '주소'로 변경, '광역시/도' 컬럼 추가

```python
df.rename(columns = {'소재지':'주소', '자치구':'시/군/구'}, inplace = True) 
df = df.drop(['Unnamed: 0', '연번'], axis=1)
df.insert(3, '광역시/도', '서울특별시', allow_duplicates=False)
df['연락처'] = df['연락처'].fillna('불명')
df
```

초기 데이터를 추출

```python
df.to_json("filename.json", orient='table', force_ascii=False)
```
